```python
class Preprocess():
    def __init__(self, text, lower=False, stem=True, sw=stopwords.words('english')):
        self.text = text
        self.lower = lower
        self.stem = stem
        self.sw = sw
    
    def clean_text(self):
        def remove_sw(string):
            list_words = string.split()
            kept_words = [word for word in list_words if not word in self.sw]
            return ' '.join(kept_words)
                
        if self.lower:
            self.text = self.text.str.lower()
        if self.sw:
            self.text = self.text.apply(remove_sw)
        self.vectorizer = TfidfVectorizer()
        self.df_dense =self.vectorizer.fit_transform(self.text)
        
    def make_df(self):
        df = pd.DataFrame(columns = self.vectorizer.get_feature_names(),
                          data = self.df_dense.toarray())
```

```python
docs = Preprocess(df['QUOTE'], stem=False, lower=True)
docs.text
docs.lower
docs.sw
docs.clean_text()
docs.text
docs.vectorizer.get_feature_names()
docs
```
